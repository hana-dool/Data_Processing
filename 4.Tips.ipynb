```python
import numpy as np
import pandas as pd

from seaborn import load_dataset
diamonds=load_dataset('diamonds')
diamonds.head()
```

# numeric type 만 선택하기

```python
df = diamonds.copy()
```

```python
df.select_dtypes(include=np.number)
```

# numeric type 이 아닌것만 선택하기

```python
df.select_dtypes(exclude=np.number)
```

# nan 이 존재하는 Row들 선택

- any method 를 써서, 하나의 null 이라도 존재하는 row 를 출력한다.

```python
df = titanic.copy()
```

```python
df[df.isnull().any(axis=1)] # isnull 을 any 와 함께 써서 출력
```

```python
# 또는 아래처럼 특정 columns에 대해서 nan 이 존재하는 df를 출력할 수도 있다.
df[df[['age','embark_town']].isnull().any(axis=1)]
```

# 특정 col 에 대해 nan 이 존재하면 drop

```python
# class,who col 에 대해 nan 이 존재하는 경우 drop
df.dropna(subset=['class','who'])
```

# 특정 col 조합에 대해 unique 한 조합만 남김

- default 는 first 값으로, 제일 처음 만나게 되는 값만 남긴다.

```python
df.drop_duplicates(subset=['survived','pclass'])
```

 # 특정 조합의 sample 이 몇개 존재하는지

- class 가 얼마나 쓰였는지에 대해서 옆에 붙여보자.

```python
df['class_num'] = df.groupby('class')['class'].transform('count')
df
```

#  sample 의 변수값 한번에 바꾸기

- Replace 와 Unique 를 같이 쓰게 된다면, 번거롭게 변수의 모든 value 들을 일일히 바꿀 필요가 없다.
- df 의 class 를 Third , Second , First -> 삼,이,일 로 바꾸고 싶다고 하자.

```python
df['class'].unique()
```

```python
# replace 를 아래와 같이 .replace([원래 values],[바꾸고 난 뒤 values]) 로 하면 바뀌게 된다.
df['class'] = df['class'].replace(df['class'].unique(),['삼','일','이'])
df 
```

# 범위를 카테고리로 바꾸기

- pd.cut 함수를 이용하자. 
- 아래와 같이 분절한다면, -inf ~ 10 을 하층민/ 10 ~ 100 을 중산층, 100~ inf 를 상류층으로 구분하게 된다.

```python
pd.cut(df['age'],[-np.inf,10,100,np.inf],labels = ['하층','중산층','상류'])
```
